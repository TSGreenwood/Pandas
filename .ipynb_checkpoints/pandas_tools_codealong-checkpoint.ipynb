## DataFrame.applymap() and Series.map()

The ```.applymap()``` method takes a function as input that it will then apply to every entry in the dataframe.

````python
def successor(x):
    return x + 1
````

````python
uci.applymap(successor).head()
````

````python
def smiley_virus(x):
    return "don't tell me what to do"
````

````python
uci.applymap(smiley_virus).head()
````

The .map() method takes a function as input that it will then apply to every entry in the Series.

````python
uci['age'].map(successor).tail(10)
````

### Anonymous Functions (Lambda Abstraction)

Simple functions can be defined right in the function call. This is called 'lambda abstraction'; the function thus defined has no name and hence is "anonymous".

````python
uci['oldpeak'].map(lambda x: round(x))[:4]
````

$\bf{\rightarrow Exercise: Use\ an\ anonymous\ function\ to\ turn\ the\ entries\ in\ age\ to\ strings}$

````python
# Your code here!
uci["age"].map(lambda a: str(a))[:3]
````

## Methods for Re-Organizing DataFrames: .groupby() and .pivot()

Those of you familiar with SQL have probably used the GROUP BY command. Pandas has this, too.

The .groupby() method is especially useful for aggregate functions applied to the data grouped in particular ways.

````python
uci.groupby('sex')
````

### .groups and .get_group()

````python
uci.groupby('sex').groups #This returns the indices corresponding to the 2 groups
````

````python
uci.groupby('sex').get_group(0).tail()
````

### Aggregating

````python
uci.groupby('sex').std()
````

$\bf{\rightarrow Exercise: Tell\ me\ the\ average\ cholesterol\ level\ for\ those\ with\ heart\ disease.}$

````python
# Your code here!
uci.groupby("target").get_group(True)["chol"].mean()
````

### .pivot()

Those of you familiar with Excel have probably used Pivot Tables. Pandas has a similar functionality.

````python
uci.pivot(values = 'sex', columns = 'target').tail(10)
````

## Methods for Combining DataFrames: .join(), .merge(), .concat(), .melt()

### .join()

````python
toy1 = pd.DataFrame([[63, 142], [33, 47]], columns = ['age', 'HP'])
toy2 = pd.DataFrame([[63, 100], [33, 200]], columns = ['age', 'HP'])
````

````python
toy1.join(toy2.set_index('age'), on = 'age',
          lsuffix = '_A', rsuffix = '_B').head()
````

### .merge()

````python
ds_chars = pd.read_csv('ds_chars.csv', index_col = 0)
````

````python
states = pd.read_csv('states.csv', index_col = 0)
````

````python
ds_chars.merge(states, left_on='home_state', right_on = 'state',
               how = 'inner')
````
